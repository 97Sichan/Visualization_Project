### 년간 고속도로 총 이용수를 구하기

```python
df.loc['고속도로 이용 누적 교통량']=df.sum(axis=0)
df
```

```python
df_1 = df.drop(['구분','차종'],axis=1)
df_1
```

```python
df_2 = df_1.transpose()
df_2
```

```python
df_2.index.name="년도"
df_2
```

```python
df_2 = df_2.reset_index()
df_2
```

```python
df_3 = df_2[['년도','고속도로 이용 누적 교통량']]
df_3
```

##### 고속도로 이용 누적 
- 2016년도 부터 2019년 까지 고속도로 이용 누적은 계속 증가하는 추세이다. 
- 하지만 2020년에 갑작스러운 감소 추세가 보인다. 
- 이는, Covid-19 의 영향을 받았다는 것을 알 수 있다. 
- 뿐만 아니라, 전기 자동차 이용량과, 전기 자동차 보조금 등과 관련되게 좀 더 고민을 해봐야 될 거 같습니다. 

```python
plt.figure(figsize=(10,8))
plt.title('년도별 고속도로 누적 교통량')

plt.xlabel('년도')

plt.ylabel('고속도로 이용 누적 교통량')

plt.plot(df_3['년도'], df_3['고속도로 이용 누적 교통량'])

colors = ['darkgrey','pink','skyblue','sandybrown','red']
plt.bar(df_3['년도'], df_3['고속도로 이용 누적 교통량'],color=colors)

plt.show()
```
