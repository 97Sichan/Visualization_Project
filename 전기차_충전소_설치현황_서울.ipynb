```python
df = pd.read_csv('./data/전기차 충전소 설치현황_20220812_서울(민지님 편집본).csv',encoding='UTF-8',low_memory=False,skiprows=2)
df.head()
```

#### 컬럼 '1' 삭제

```python
del df['1']
df
```

#### '구' 컬럼 추가

```python
df['주소'] = df['주소'].astype('str')
```

```python
df_1 = df['주소'].str.split(' ')
df_1
```

```python
df['구'] = df_1.str.get(1)
df
```

#### 컬럼 순서 변경
- 지원차종까지 하면 너무 복잡할 거 같다는 생각에 임의로 뺐습니다.

```python
df = df[['시구','설치장소','주소','구','급속충전기(대)','완속충전기(대)']]
df
```

### 구별 급속충전기, 완속충전기 수

```python
test_df1 = pd.pivot_table(df,
                   index='구',
                   values=['급속충전기(대)','완속충전기(대)'],
                   aggfunc='sum')
test_df1
```

```python
test_df1.to_csv('./data/charger.csv')
```

```python
#한국 지도 데이터 이용해서 단계 구분도 그리기
import json
#한국지도 경계의 json data 파일 가져오기
#한글이 포함된경우에는 그냥 path만 연결하면 인코딩 문제 발생
#json.load( encoding='utf-8') 이용해서 변수에 data 저장 후 사용하는 것이 일반적

geo_path='./data/02. skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path,encoding='utf-8'))
# geo_str

#서울을 중심으로 기본 지도 출력
#위경도[37.5502,126.982]
map=folium.Map(location=[37.5502,126.982],
               zoom_start=11,
               tiles='Stamen Toner')
map

#단계 구분도 data 읽어오기
#csv 읽어오기 : pd.read_csv(경로포함파일명,encoding=,index_col='지정인덱스 필드명')
f_test=pd.read_csv('./data/charger.csv',encoding='utf-8',index_col='구')
f_test
```

### 급속충전기(대)

```python
map.choropleth(geo_data=geo_str,
              data = f_test,
              columns=[f_test.index,'급속충전기(대)'],
              key_on = 'feature.id',
              fill_color='PuRd',
              legend_name='급속충전기')
folium.LayerControl().add_to(map)
map
```

### 완속충전기(대)

```python
map.choropleth(geo_data=geo_str,
              data = f_test,
              columns=[f_test.index,'완속충전기(대)'],
              key_on = 'feature.id',
              fill_color='PuRd',
              legend_name='완속충전기')
folium.LayerControl().add_to(map)
map
```
